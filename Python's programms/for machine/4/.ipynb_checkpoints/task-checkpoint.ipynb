```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

# Экспоненциальное распределение с плотностью $p(x) = \lambda e^{-\lambda x}\cdot I(x\geq 0)$ с параметром $\lambda = 1$.
Выборка размера 1000 из экспоненциального распределения.
Гисторгамма плотности распределения и сама плотность изображены на графике.

```python
exp_d = sts.expon(scale = 1)
sample = exp_d.rvs(1000)
plt.hist(sample, bins = 25, normed = True)

x = np.linspace(0,10,100)
pdf = exp_d.pdf(x)
plt.plot(x, pdf)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')

plt.show()
```

По ЦПТ: $\bar{X} \sim N(EX_{1}^{}, \frac{DX_{1}^{}}{n})$, где $n$ - размер выборки.

$EX_{1}^{} = \frac{1}{\lambda} = 1$

$ \frac{DX_{1}^{} }{n} = \frac{1}{\lambda^{2} \cdot n} = \frac{1}{n}$

## Гистограмма для средних значений на выборке размера 5

$N(1; 0.2)$

```python
sample = np.array([])
for x in xrange(1000):
    sample = np.append(sample, np.average(exp_d.rvs(5)))
```

```python
plt.hist(sample, bins = 20, normed = True)

x = np.linspace(0,5,1000)
norm_rv = sts.norm(1, np.sqrt(0.2))
pdf = norm_rv.pdf(x)
plt.plot(x, pdf)

plt.xlim(0,3)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')

plt.show()
```

## Гистограмма для средних значений на выборке размера 10

$N(1; 0.1)$

```python
sample = np.array([])
for x in xrange(1000):
    sample = np.append(sample, np.average(exp_d.rvs(10)))
    
plt.hist(sample, bins = 20, normed = True)

x = np.linspace(0,5,1000)
norm_rv = sts.norm(1, np.sqrt(0.1))
pdf = norm_rv.pdf(x)
plt.plot(x, pdf)

plt.xlim(0,3)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')

plt.show()
```

$N(1; 0.02)$

```python
sample = np.array([])
for x in xrange(1000):
    sample = np.append(sample, np.average(exp_d.rvs(50)))
    
plt.hist(sample, bins = 20, normed = True)

x = np.linspace(0,5,1000)
norm_rv = sts.norm(1, np.sqrt(0.02))
pdf = norm_rv.pdf(x)
plt.plot(x, pdf)

plt.xlim(0,2)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')

plt.show()
```

# Вывод

С ростом числа n точность аппроксимации распределения выборочных средних нормальным растет. При n = 5 гистограмма лишь отдаленно напоминает вид нормального распределения, при n = 10 принимает похожую форму, а при n = 50 отличия незначительны.
