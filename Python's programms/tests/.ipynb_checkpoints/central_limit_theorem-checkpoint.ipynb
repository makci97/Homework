```python
import scipy.stats as sts
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

Create Pareto distribution with parameter b = 10 and generate a sample:

```python
b = 10
pareto_rv = sts.pareto(b)
sample = pareto_rv.rvs(size=1000)
```

Create a histogram of the sample from Pareto distribution:

```python
plt.hist(sample, bins=20, normed=True)
plt.ylabel('number of samples')
plt.xlabel('$x$')
```

Comparison of the histogram and theoretical Pareto pdf:

```python
x = np.linspace(0.5, 2.5, 100)
pdf = pareto_rv.pdf(x)
plt.hist(sample, bins=20, normed=True, label='sample histogram')
plt.plot(x, pdf,'r-', lw=2, label='pareto pdf')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

Central limit theorem for samples of size 5:

```python
samples = []
n = 5
for j in range(0, 5000):
    samples.append(pareto_rv.rvs(size=n))
averages = []
for s in samples:
    averages.append(np.average(s))
approx = sts.norm(pareto_rv.mean(), np.sqrt(pareto_rv.var()/n))
x = np.linspace(1, 1.4, 100)
pdf = approx.pdf(x)
plt.plot(x, pdf,'r-', label='normal distribution pdf', lw=2)
plt.hist(averages, bins=20, normed=True, label='sample means')
plt.legend()
plt.xlabel('sample\'s average')
plt.ylabel('$f(x)$')
```

```python
samples = []
n = 10
for j in range(0, 1000):
    samples.append(pareto_rv.rvs(size=n))
averages = []
for s in samples:
    averages.append(np.average(s))
approx = sts.norm(pareto_rv.mean(), np.sqrt(pareto_rv.var()/n))
x = np.linspace(1, 1.4, 100)
pdf = approx.pdf(x)
plt.plot(x, pdf,'r-', label='normal distribution pdf', lw=2)
plt.hist(averages, bins=20, normed=True, label='sample means')
plt.legend()
plt.xlabel('sample\'s average')
plt.ylabel('$f(x)$')
```

Central limit theorem for samples of size 50:

```python
samples = []
n = 50
for j in range(0, 1000):
    samples.append(pareto_rv.rvs(size=n))
averages = []
for s in samples:
    averages.append(np.average(s))
approx = sts.norm(pareto_rv.mean(), np.sqrt(pareto_rv.var()/n))
x = np.linspace(1, 1.4, 100)
pdf = approx.pdf(x)
plt.plot(x, pdf,'r-', label='normal distribution pdf', lw=2)
plt.hist(averages, bins=20, normed=True, label='sample means')
plt.legend()
plt.xlabel('sample\'s average')
plt.ylabel('$f(x)$')
```

Normal distribution approximating the distribution of sample means was built with theoretical moments of Pareto distribution:

The mathematical expectation:

$E(x) =
\begin{cases}
\infty, b \le 1 \\[2ex]
\frac{bx_m}{b-1}, b > 1
\end{cases}$

The variance:

$Var(x) = 
\begin{cases}
\infty, b \in (1,2]\\[2ex]
\left(\frac{x_m}{b-1}\right)^2\cdot\frac{b}{b-1}, b > 2
\end{cases}$

In our case Xm = 1, b = 10.

# Conclusions

Distribution resembles normal distribution when <b>n</b> is high. Approximation accuracy increases with the increase of <b>n</b>.
