```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma, norm
%matplotlib inline
#задаем параметры гамма-распределения
shape, loc, scale, size = 2, 0, 2, 1000

r = gamma.rvs(shape, loc, scale, size)

x = np.linspace(0, 20, 100)
rv = gamma(shape, loc, scale)
gamma_pdf = rv.pdf(x)

gamma_mean, gamma_var = rv.stats()
```

```python
#строим гистограмму гамма распределения и график теоретической плотности распределения
plt.hist(r, normed = True, label = 'gamma hist')
plt.plot(x, gamma_pdf, label = 'theory pdf')
plt.title("Gamma Histogram")
plt.xlabel("x")
plt.ylabel("Frequency")
plt.legend()
plt.show() 
```
