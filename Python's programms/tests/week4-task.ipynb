```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

## Генерируем выборку объема 1000, строим гистограмму и теоретическую плотность

```python
c = 4.07
weib_rv = sts.dweibull(c)
sample = weib_rv.rvs(1000)
plt.hist(sample, normed = True, label = 'Hist double Weibull',bins = 17)
plt.plot()

x = np.linspace(weib_rv.ppf(0.001), weib_rv.ppf(0.999), 1000)
y = sts.dweibull.pdf(x, c)
plt.plot(x, y, label = 'Double Weibull')
plt.legend(loc = 'lower right')
```

## Показываем параметры исходноо распределения и параметры подвыборок объема n = 5, 10, 50 элементов

```python
print('Параметры распределения')
print('мат.ожидание:', weib_rv.mean(), ' дисперсия:', weib_rv.std() ** 2)


for n in [5, 10, 50]:
    m = 1000
    
    smp_mtr = np.random.randn(m, n)
    mean_smp = np.random.randn(m)
    disp_smp = np.random.randn(m)
    for i in range(0,m):
        smp_mtr[i, :] = weib_rv.rvs(n)
        mean_smp[i] = np.mean(smp_mtr[i, :])
        m = np.mean(mean_smp[i])
        disp_smp[i] = np.mean((smp_mtr[i, :] - m) ** 2)
    f, ax1 = plt.subplots(1, 1, sharex=True)
    lbl = 'Subset of ' + str(n) + ' elements'
    x = np.linspace(min(mean_smp),max(mean_smp),100)
    norm_rv = sts.norm(weib_rv.mean(), weib_rv.std() / np.sqrt(n))
    y = norm_rv.pdf(x)
    ax1.hist(mean_smp, normed = True, bins = 15, label = lbl)
    ax1.plot(x, y, label = 'theoretical')
    plt.ylabel('number of samples, $f(x)$')
    plt.xlabel('$x$')
    ax1.legend(loc = 'lower right')
    print('для n = ',n,'мат.ожидание:', round(np.mean(mean_smp),10), 'дисперсия:', round(np.mean(disp_smp),10))
```

### При увеличении числа элементов в подвыборке выборочные средние и дисперсии приближаются к теоретическим.
